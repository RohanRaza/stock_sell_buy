```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stock = pd.read_csv("stock.csv", parse_dates =["Date"])
stock
```

```python
data = pd.DataFrame()
```

```python
data['Date'] = stock['Date']
data['Apple'] = stock['Adj Close Price']
```

```python
data
```

```python
data['Apple'].head(30).mean(),data['Apple'].iloc[1:30].mean()
```

#### We will use Rolling method to find  mean of every 30 values

```python
data['sma30'] =data['Apple'].rolling(30).mean()
```

```python
pd.set_option('display.max_rows',None)
```

```python
data
```

```python
data['sma100'] = data['Apple'].rolling(100).mean()
```

```python
data.head(110)
```

#### Now we have to make function that helps when we should buy or sell stocks

```python
def buy_sell(df):
    buy_stock = [] # when 30 days mean > 100 days mean
    sell_stock =[] # when 30 days mean < 100 days mean
    hold_stock = [] # 1 for hold, 0 for not hold
    action = [] # 1 for buy , -1 for sell, 0 for hold
    sig = 0
    
    for i, row in df.iterrows():
        if row['sma30'] > row['sma100']: # buy
            if sig != 1:
                buy_stock.append(row['Apple'])
                sell_stock.append(np.nan)
                action.append(1)
                hold_stock.append(1)
                sig = 1
            else:
                buy_stock.append(np.nan)
                sell_stock.append(np.nan)
                action.append(0)
                hold_stock.append(1)
        elif row['sma30'] < row['sma100']: # sell
            if sig != -1:
                buy_stock.append(np.nan)
                sell_stock.append(row['Apple'])
                action.append(-1)
                hold_stock.append(0)
            else:
                buy_stock.append(np.nan)
                sell_stock.append(np.nan)
                action.append(0)
                hold_stock.append(0)
        else:
            buy_stock.append(np.nan)
            sell_stock.append(np.nan)
            action.append(0)
            hold_stock.append(0)
            
    return buy_stock,sell_stock,hold_stock,action
            
```

```python
stocks = buy_sell(data)
```

```python
data['Buy_Stock'],data['Sell_Stock'],data['Hold_Stock'],data['Action'] = stocks
```

```python
plt.figure(figsize=(14,6))
plt.style.use('dark_background')

plt.plot(data['Date'],data['Apple'],label = 'Apple',alpha = 0.6)
plt.plot(data['Date'],data['sma30'],label = 'sma30',alpha = 0.9,color='blue')
plt.plot(data['Date'],data['sma100'],label = 'sma100',alpha = 0.9,color ='yellow')


#scater
plt.scatter(data['Date'],data['Buy_Stock'],alpha = 0.9,marker = '^',color = 'white')
plt.scatter(data['Date'],data['Sell_Stock'],alpha = 0.9,marker = 'v',color = 'red')

plt.title("Apple Historical Adj Close Price")
plt.xlabel("From May, 2014 to 22 May, 2020")
plt.ylabel("Adj Close Price")

plt.show()
```
